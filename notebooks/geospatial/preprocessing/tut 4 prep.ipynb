```python
import requests
import lxml.html as lh
import pandas as pd
```

```python
url='https://www.timeshighereducation.com/student/best-universities/best-universities-europe#survey-answer'

# Create a handle, page, to handle the contents of the website
page = requests.get(url)

# Store the contents of the website under doc
doc = lh.fromstring(page.content)

# Parse data that are stored between <tr>..</tr> of HTML
t_elements = doc.xpath('//tr')
```

```python
print(len(t_elements))
print(t_elements[464].text_content())
```

```python
col=[]

for j in range(1,464):
    T=t_elements[j]
    
    # i is the index of our column
    i=-1
    
    # Iterate through each element of the row
    for t in T.iterchildren():
        data=t.text_content() 
        if i==1:
            col.append(data)
      #Increment i for the next column
        i+=1
```

```python
from geopandas.tools import geocode

def get_location(i):
    try: 
        return geocode(i).geometry.values[0]
    except:
        return float('NaN')
```

```python
from tqdm import tqdm

locations = [get_location(i) for i in tqdm(col[:100])]
```

```python
zippedList =  list(zip(col, locations))
df = pd.DataFrame(zippedList, columns = ['Name' , 'Location']) 
df
```
